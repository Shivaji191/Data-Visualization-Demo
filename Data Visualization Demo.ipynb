## Data Visualization

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
h=sns.load_dataset("healthexp")
h
```

```python
sns.lineplot(x="Country",y="Life_Expectancy",data=h,size=15,)
plt.show()
```

```python
sns.lineplot(x="Country",y="Spending_USD",data=h,)
plt.title("USD DATA")
plt.show()
```

```python
sns.barplot(x="Country",y="Life_Expectancy",data=h)
plt.show()
```

```python
sns.boxplot(x="Country",y="Life_Expectancy",data=h)
```

```python
sns.scatterplot(x="Country",y="Life_Expectancy",data=h)
```

```python
sns.violinplot(x="Country",y="Life_Expectancy",data=h)
```

```python
sns.stripplot(x="Country",y="Life_Expectancy",data=h)
```

```python
sns.swarmplot(x="Country",y="Life_Expectancy",data=h)
```

```python
sns.histplot(x="Country",y="Life_Expectancy",data=h)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```
